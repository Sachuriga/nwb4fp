```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'/Volumes/ntnu/mh-kin/quattrocolo/crhip/Sachuriga/Ephys_Recording/CR_CA1/63383/63383_2024-07-28_unitmatchResults/MatchTable.csv')
df1 = df[df['Matches']==1]
df2 = df1[df1['UM Probabilities']>0.95]

sesion_num = 1
df_s = df2[(df2['RecSes 1'] == sesion_num) & (df2['RecSes 2'] != sesion_num)]
df_s
```

```python
all_temp_rows = []
sesion_num = 1
df_s = df2[(df2['RecSes 1'] == sesion_num) & (df2['RecSes 2'] != sesion_num)]
match_sessions = np.unique(df_s['RecSes 2'])
for s in match_sessions:
    temp_df = df_s[df_s['RecSes 2']==s]
    for i in np.unique(temp_df['ID1']):
        temp_id_df = temp_df[temp_df['ID1']==i].sort_values(by='TotalScore',ascending=False)
        temp_row = temp_id_df.head(1)
        all_temp_rows.append(temp_row)

if len(match_sessions)>=2:
    print("here")
    df_s2 = df2[(df2['RecSes 1'] == match_sessions[0]) & (df2['RecSes 2'] == match_sessions[1])]
    for j in np.unique(df_s2['ID1']):
        temp_id_dfs2 = df_s2[df_s2['ID1']==j].sort_values(by='TotalScore',ascending=False)
        temp_rows = temp_id_dfs2.head(1)
        all_temp_rows.append(temp_rows)

new_df = pd.concat(all_temp_rows, ignore_index=True)
```
